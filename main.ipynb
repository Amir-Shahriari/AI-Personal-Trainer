```python
from fastapi import FastAPI
import pandas as pd

# Load yoga exercise data
yoga_data = pd.read_csv("yoga_data.csv")

# Initialize FastAPI app
app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Welcome to the Yoga Session Planner"}

@app.post("/plan/")
def create_yoga_plan(duration: int, intensity: str, muscle_groups: list[str]):
    """
    Generate a yoga session plan based on the requested duration, intensity, and muscle groups.
    """
    # Filter exercises by intensity and muscle groups
    filtered_data = yoga_data[
        (yoga_data["intensity"].str.contains(intensity, case=False)) &
        (yoga_data["muscle_groups"].apply(lambda x: any(muscle in x for muscle in muscle_groups)))
    ]

    if filtered_data.empty:
        return {"message": "No matching exercises found. Try adjusting the inputs."}

    # Calculate total session plan
    session_plan = []
    total_duration = 0
    for _, row in filtered_data.iterrows():
        if total_duration >= duration:
            break
        pose_duration = int(row["duration"].split()[0])  # Simplify duration handling
        total_duration += pose_duration
        session_plan.append({
            "pose": row["pose"],
            "instructions": row["instructions"],
            "duration": row["duration"]
        })

    return {"total_duration": total_duration, "plan": session_plan}
```
